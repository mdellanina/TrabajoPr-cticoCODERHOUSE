## 1.3 DataSet

```python
path = "https://github.com/mdellanina/TrabajoPracticoCODERHOUSE/raw/main/SolarPrediction/SolarPrediction.csv"
df = pd.read_csv(path)
display(df)
```

A este DataSet transformado en dataFrame, le haremos algunas modificaciones ya que vemos que es necesario acomodar los datos para que se puedan analizar mejor. Las modificaciones propuestas son las siguientes:

*   El Campo "Data" se ajusta solo a la fecha, se sacará la hora, ya que ese dato esta en el campo "Time"
*   El campo UNIXTime se elimina, ya que es redundante (Solo es una ref de fecha global)
*   El Campo "Data" estaba en formato dateTime y se cambia a Object para que la tome como categorica.
*   Agregar una columna nueva con un string diferente representando los 4 puntos cardinales principales, a partir de calcular si la variable "WindDirección(Degrees)" varia entre 4 diferentes rangos de angulos.
*   Reemplazar el valor de "Temperature" que esta medida en °F a °C (unidad de medida en Argentina)
*   Reemplazar el valor de "Speed" que esta medida en mph a km/h, y generar a partir de este valor actualizado, una variable categorica nueva con un string diferente representando los 5 denominaciones de vientos, a partir de calcular si la variable "Speed" varia entre 5 diferentes rangos de velocidades.



```python
# El campo UNIXTime podria eliminarse, ya que es redundante (Solo es una ref de fecha global)
df = df.drop(['UNIXTime'], axis=1)
# El Campo "Data" estaba en formato dateTime y se cambia a Object para que la tome como categorica.
df['Data']=pd.to_datetime(df['Data'], format='%m/%d/%Y %I:%M:%S %p')
df['Data'] = df['Data'].astype(str)
# Agregamos una columna nueva con un string diferente representando los 4 puntos cardinales principales, a partir de calcular si la variable "WindDirección(Degrees)" varia entre 4 diferentes rangos de angulos.

def categorize_wind_direction(row):
    direction = row["WindDirection(Degrees)"]
    if (0 <= direction < 45) or (315 <= direction <= 360) :
        return "Norte"
    elif 45 < direction <= 135:
        return "Este"
    elif 135 < direction <= 225:
        return "Sur"
    else:
        return "Oeste"
df["WindDirectionCategory"] = df.apply(categorize_wind_direction, axis=1)
# reemplazar el valor de "Temperature" que esta medida en °F a °C
df['Temperature'] = (df['Temperature'] - 32) * 5/9
# reemplazar el valor de "Speed" que esta medida en mph a km/h
df['Speed'] = df['Speed'] * 1.609344

def categorize_wind_speed(row):
    speed = row["Speed"]
    if 1 <= speed <= 11:
        return "Brisa Suave"
    elif 11 < speed <= 28:
        return "Leve a moderado"
    elif 28 < speed <= 48:
        return "Regulares a Fuertes"
    elif 48 < speed <= 73:
        return "Muy Fuertes o Temporales"
    else:
        return "Temporal muy fuerte o Huracan"

df["WindSpeedCategory"] = df.apply(categorize_wind_speed, axis=1)
display(df)
```

## 3.3 Analisis de variables - Agrupación Diaria

En esta sección, agrupamos los datos por día, con la siguiente selección de dato para cada variable:

### 4.3.1 Modelado univariado "Radiation vs temperature"

En primer lugar veremos si el modelo de regresión Lineal nos brinda una solución que ajuste y tenga un buen score.

```python
#Generamos un df_new_uv (univariado)
df_new_uv = df[['Temperature', 'Radiation']]
```

```python
#Calcular el coeficiente de correlación entre X e Y
corr_coeff =df_new_uv['Radiation'].corr(df_new_uv['Temperature'], method='pearson')

# Imprimir el coeficiente de correlación

print("Coeficiente de correlación de Pearson: ", corr_coeff)
```

```python
#Feature/variable independiente
X = df_new_uv[['Temperature']]

#Target/variable dependiente (Variable a predecir)
y = df_new_uv[['Radiation']]
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state=0)

# nos podemos quedar con el 70% para entrenar y 30% para testear

#Entrenamos el modelo con los datos
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Obtener la pendiente
slope = regressor.coef_

# Obtener los coeficientes de intercepción
intercept = regressor.intercept_

print("este es el valor de la pendiente:", slope.round(2))
print("este es el valor del interceptor:", intercept.round(1))
print("La ecuación de la recta de mejor ajuste es--> y = ", float(slope.round(2)), "* X +", float(intercept.round(2)))

```

### 4.3.2 Modelado Lineal Multivariado

```python
# Import the necessary libraries
from sklearn.linear_model import LinearRegression
df_new_mv = df[['Radiation', 'Temperature', 'Pressure','Humidity', 'WindDirection(Degrees)', 'Speed']]
# Separate the features and target variable
X = df_new_mv[['Temperature', 'Pressure','Humidity', 'WindDirection(Degrees)', 'Speed']]
y = df_new_mv['Radiation']

# nos podemos quedar con el 70% para entrenar y 30% para testear
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state=0)

# Create a linear regression model
regressor = LinearRegression()

# Train the model
regressor.fit(X_train, y_train)

# Get the coefficients and intercept
coefficients = regressor.coef_
intercept = regressor.intercept_

# Print the results
print('Coefficients:', coefficients)
print('Intercept:', intercept)
```

```python
#score de Test
regressor.score(X_test, y_test)
```

```python
#score de Train

regressor.score(X_train, y_train)
```

El score del modelo multivariado dio un valor bajo tambien, por lo que buscaremos algunas respuestas con regresión logica usando "DecisionTree"

### 4.3.3 Modelado Multivariado con "DecisionTreeRegressor"

```python
from sklearn.tree import DecisionTreeRegressor

tree = DecisionTreeRegressor(random_state=10)
tree.fit(X_train, y_train)
tree.score(X_train, y_train)
```

```python
tree.score(X_test, y_test)
```

**Interpretación:**

El hecho de obtener un R2 de 0.99 en los datos de entrenamiento (entrenamiento) y 0.51 en los datos de prueba (test) indica que el modelo de regresión de árbol de decisiones muestra signos de sobreajuste (overfitting).

Es recomendable verificar si el modelo está sobreajustando los datos de entrenamiento. Un R2 extremadamente alto puede indicar sobreajuste, donde el modelo se ajusta demasiado bien a los datos específicos de entrenamiento pero no generaliza bien a datos nuevos.

**Recomendación:**

Utilizar técnicas como la validación cruzada o la separación adecuada de datos de entrenamiento y prueba para evaluar de manera más completa el rendimiento del modelo. Esto requiere de readecuar lo datos y analizarlos de nuevo.
Otro camino para abordar el problema de sobreajuste experimentado con el modelo es modificar algunos de los hiperparámetros, por ejemplo:

**Metricas:**

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error, r2_score
# Predecir con X_test
y_pred = regressor.predict(X_test)
# Predecir con X_train
y_pred_train = regressor.predict(X_train)
# Calcular el error absoluto medio (MAE)
mae = mean_absolute_error(y_train, y_pred_train)
print("Error absoluto medio (MAE) en train es:", mae)
r2 = r2_score(y_test,y_pred)
print("El r2 en train es: ",r2)
mae = mean_absolute_error(y_test, y_pred)
print("Error absoluto medio (MAE) en test es:", mae)
r2 = r2_score(y_test,y_pred)
print("El r2 en test es: ",r2)
```

Estos valores nos dicen que:

*   Esperariamos que el MAE en test sea ligeramente mayor que el MAE en train debido a que el modelo no ha visto los datos de prueba durante el entrenamiento. Sin embargo, la diferencia es bastante pequeña, lo cual indica que el modelo generaliza de manera razonable bien a datos nuevos.
*   Un MAE de alrededor de 157 unidades parece razonable para el contexto del problema y parece aceptable



```python
from sklearn.metrics import mean_squared_error
print("MSE",mean_squared_error(y_test,y_pred))
```

No parece ser robusto a valores atípicos, lo cual parece ser algo común a este dataset.

```python
# Definir el modelo con ajuste de hiperparámetros
tree = DecisionTreeRegressor(random_state=10, max_depth=10, min_samples_split=10, min_samples_leaf=5)
tree.fit(X_train, y_train)
print("Score Train: ",tree.score(X_train, y_train))
print("Score Test: ",tree.score(X_test, y_test))
```

los puntajes obtenidos después de ajustar los hiperparámetros muestran una mejora en la capacidad del modelo para generalizar a datos nuevos, lo cual es un paso importante en la dirección correcta.

```python
# Predecir con X_test
y_pred = regressor.predict(X_test)
# Predecir con X_train
y_pred_train = regressor.predict(X_train)
# Calcular el error absoluto medio (MAE)
mae = mean_absolute_error(y_train, y_pred_train)
print("Error absoluto medio (MAE) en train es:", mae)
r2 = r2_score(y_test,y_pred)
print("El r2 en train es: ",r2)
mae = mean_absolute_error(y_test, y_pred)
print("Error absoluto medio (MAE) en test es:", mae)
r2 = r2_score(y_test,y_pred)
print("El r2 en test es: ",r2)
```

Añadimos a este analisis Vaidación Crizada, que podría mejorar el modelo y para evaluar el rendimiento del modelo de manera más robusta. Esto deberia mejorar la estimación del rendimiento del modelo en datos no vistos.

```python
from sklearn.model_selection import cross_val_score

# Ejemplo de validación cruzada con 5 folds
scores = cross_val_score(tree, X_train, y_train, cv=5)
print("Mean Cross-Validation R^2:", scores.mean())
test_score = tree.score(X_test, y_test)
print("Test R^2:", test_score)
```

Interpretacion:

*   La mejora en el R2 sugiere que el modelo ajustado con validación cruzada generaliza mejor a datos nuevos. Esto indicaría que el modelo original podría haber estado sobreajustado (overfitting) a los datos de entrenamiento antes de aplicar la validación cruzada, lo cual habiamos supuesto anteriormente.
*   Indica que el modelo tiene una buena capacidad de generalización y puede hacer predicciones más precisas en situaciones del mundo real donde se encuentren nuevos datos.

En resumen, la mejora de 0.56 a 0.67 en el R2 con la validación cruzada significa que el modelo ahora es más confiable y efectivo en la predicción de nuevos, lo cual es un resultado muy positivo.

Sería deseable obtener un R2 mas alto aún, quizas requiera manipular los datos de entrada de manera de optimizarlos, o bien revisar los ajustes de hiperparametros o utilizar otros metodos de ajuste.
